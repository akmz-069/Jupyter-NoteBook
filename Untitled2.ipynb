```python


import pandas as pd
import seaborn as sb
import matplotlib.pyplot as mtb
import numpy as np
from sklearn.cluster import KMeans

np.random.seed(0)
num_points = 300
data = np.random.rand(num_points, 2)

group_size = num_points // 3
group1 = data[:group_size]
group2 = data[group_size:2*group_size]
group3 = data[2*group_size:]

offset1 = np.array([1, 1])
offset2 = np.array([5, 5])
offset3 = np.array([9, 1])


group1 = group1 + offset1
group2 = group2 + offset2
group3 = group3 + offset3


data = np.concatenate((group1, group2, group3))
# print(data.head(10))

df = pd.DataFrame(data, columns=['x', 'y'])

kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(df[['x', 'y']])

sb.scatterplot(data=df, x='x', y='y', hue='cluster', palette='viridis')
mtb.title('KMeans Clustering of Random Data Points')
mtb.show()
```
