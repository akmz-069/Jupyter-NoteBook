```python
#example workflow
%matplotlib inline 
import matplotlib.pyplot as plt
#data
x=[1,2,3,4]
y=[11,22,33,44]
#setup plot
fig ,ax=plt.subplots(figsize=(10,10))
#data plot
ax.plot(x,y)
#plot customize
ax.set(title="example-plot",
      xlabel="x-axis",
      ylabel="y-axis")
#save the figure
fig.savefig("images/example-plot.png")
plt.show()
```
