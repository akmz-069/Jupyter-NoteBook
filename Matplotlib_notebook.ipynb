### Plotting from pandas dataframes 

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\\$\\,\\.]','',regex=True)
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales Date"]=pd.date_range("1/1/2024",periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="Sales Date", y="Total Sales")
plt.show()
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter")
plt.show()
```

```python
#another csv
heart_disease=pd.read_csv("dataset.csv")
heart_disease.head(10)
```

```python
heart_disease["age"].plot.hist(bins=50)
plt.show()
```

```python
heart_disease.plot.hist(figsize=(10,40),subplots=True);
plt.show()
```

```python
age_over_50=heart_disease[heart_disease["age"]>50]
age_over_50.head(15)
```

```python
age_over_50.plot(kind="scatter",
                x="age",
                 y="chol",
                c="target");
plt.show()
```

```python
fig, (ax0, ax1) =plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True )
scatter=ax0.scatter(
                x=age_over_50["age"],
                y=age_over_50["chol"],
                c=age_over_50["target"],
                );
ax0.set(title="Heart Disease and Cholestrol Levels",
      ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="Target")

#mean
ax0.axhline(y= age_over_50["chol"].mean(),linestyle='--')
#add data 
scatter=ax1.scatter(x=age_over_50["age"],
                y=age_over_50["thalach"],
                c=age_over_50["target"])
#customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="age",
       ylabel="Max Heart Rate")
#legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")
#add meanline 
ax1.axhline(y= age_over_50["thalach"].mean(),linestyle='--')

plt.show()
```

```python
plt.style.use('ggplot')
fig, (ax0, ax1) =plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True )
scatter=ax0.scatter(
                x=age_over_50["age"],
                y=age_over_50["chol"],
                c=age_over_50["target"],
                cmap="plasma");
ax0.set(title="Heart Disease and Cholestrol Levels",
      ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="Target")

#mean
ax0.axhline(y= age_over_50["chol"].mean(),linestyle='--')
#add data 
scatter=ax1.scatter(x=age_over_50["age"],
                y=age_over_50["thalach"],
                c=age_over_50["target"],
                   cmap="winter")
#customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="age",
       ylabel="Max Heart Rate")
#legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")
#add meanline 
ax1.axhline(y= age_over_50["thalach"].mean(),linestyle='--')

plt.show()
```
