```python
#example workflow
%matplotlib inline 
import matplotlib.pyplot as plt
#data
x=[1,2,3,4]
y=[11,22,33,44]
#setup plot
fig ,ax=plt.subplots(figsize=(10,10))
#data plot
ax.plot(x,y)
#plot customize
ax.set(title="example-plot",
      xlabel="x-axis",
      ylabel="y-axis")
#save the figure
fig.savefig("images/example-plot.png")
plt.show()
```

### Making figures with Numpy Arrays


```python
import numpy as np
```

```python
#data
x=np.linspace(0,10,100)
x[:10]
```

```python
#plot 
fig, ax=plt.subplots()
ax.plot(x,x**2);
plt.show()
```

```python
fig, ax=plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
fig, ax=plt.subplots()
ax.scatter(x, np.sin(x));
plt.show()
```

```python
#plot from dictionary
nut_butter_prices= {"Almond butter": 10,
                   "Cashew Butter":8,
                   "Peanut Butter" : 12 
                   }
fig ,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),height = nut_butter_prices.values())
ax.set(title="Nut butter prices",
       xlabel="Nut butter ",
      ylabel="Prices $");
plt.show()
```

```python
fig, ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
plt.show()
```
